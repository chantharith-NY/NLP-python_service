```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

MODEL_PATH = r"D:\RAC-AI Projects\Khmer-Text-Full-System-Rith\NLP-python_service\models\Qwen3-4B-Khmer-text-summarizer-20260310T014415Z-1-002"

print("Loading model...")

# 4bit quantization config
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True,
    ignore_mismatched_sizes=True,
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, trust_remote_code=True)

model.eval()

print("Model loaded successfully!")
```

```python
def generate_summary(text, max_length_ratio=0.3, max_new_tokens=1024):
    """
    Generate summary with dynamic length based on input.
    For very long inputs, increase max_new_tokens up to 1024.
    """
    # Calculate target length
    input_token_count = len(tokenizer(text)["input_ids"])
    target_max_tokens = int(input_token_count * max_length_ratio)
    target_max_tokens = min(target_max_tokens, max_new_tokens)  # Cap at 1024

    # Use prompt that encourages full summarization
    prompt = f"""<|im_start|>user\nбһҹбһјбһҳбһҹбһ„бҹ’бһҒбҹҒбһ”бһўбһҸбҹ’бһҗбһ”бһ‘бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбһҮбһ¶бһўбһҸбҹ’бһҗбһ”бһ‘бһҒбҹ’бһӣбһё бһ“бһ·бһ„бһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢбҹ” бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбһҸбҹ’бһҗбһ”бһ‘бһңбһ¶бһҳбһ¶бһ“бһ—бһ¶бһ–бһңбһ·бһҹбҹ’бһҹбһҳбһҡбһ”бһҹбҹӢбһңбһ¶бһ‘бҹ… бһҹбһјбһҳбһҹбһ„бҹ’бһҒбҹҒбһ”бһҮбһ¶бһҳбһҪбһҷбһ“бһ№бһ„бһ”бҹ’бһҡбһҳбһ¶бһҺ {target_max_tokens} бһ–бһ¶бһҖбҹ’бһҷбҹ”\n{text}<|im_end|>\n<|im_start|>assistant\n"""

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, padding=True).to(
        "cuda"
    )

    outputs = model.generate(
        **inputs,
        max_new_tokens=1024,
        temperature=0.7,
        top_p=0.9,
        top_k=20,
        do_sample=True,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id,
    )

    # Decode only newly generated tokens, not the full prompt
    generated_ids = outputs[0][inputs["input_ids"].shape[1] :]
    assistant_response = tokenizer.decode(
        generated_ids, skip_special_tokens=True
    ).strip()

    return assistant_response
```
